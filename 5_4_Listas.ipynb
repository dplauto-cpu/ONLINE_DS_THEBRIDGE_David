## Listas
[al indice](#Contenidos)  
Se trata de otro de los tipos de datos de Python más usados. Dentro de las colecciones, que veremos más adelante, la lista es la colección que normalmente se le da más uso. **Nos permiten almacenar conjuntos de variables u objetos**, y son elementos de lo más versátiles puesto que podemos almacenar objetos de distintos tipos, modificarlos, eliminarlos, meter listas dentro de listas... Sus dos caractrísticas principales son:
* **Mutables**: una vez se ha creado la lista, se puede modificar
* **Ordenada**: Los elementos tienen un cierto orden, lo que nos permite acceder al elemento que queramos teniendo en cuenta tal orden

En cuanto a su sintaxis, cuando declaremos la lista simplemente hay que separar cada elemento con comas, y rodearlo todo con corchetes.

```python
# Método de lista
numeros= [1,3,5,6]
print(numeros, type(numeros))
```

```python
# Lista de Strings, se pueden hacer listas de muchas cosas, o solo de números
Strings = ["Esto", "es", "una", "lista"]
print(Strings, type(Strings))

# Lista de Booleanos
booleanos = [True, False, not False, True or False, False and False]
print(booleanos)
```

```python
# Lista de tipos mezclados
mix = ["esto","es", 12, -3.5, True, not False]
print(mix)

# También se pueden meter listas dentor de listas
listas = ["Hola", 12, 23, [23, True, False]]
print(listas) #esta lista tiene 4 elementos porque 3 de ellos forman uno solo (otra lista), así
len(listas), len (mix)
```

```python
# Las listas también se pueden concatenar
concatenacion = listas + mix
print(concatenacion)
print(type(concatenacion))
# atención, la lista es ordenada, y el orden sí que importa, no es conmutativa:
concatenacion2 = mix + listas
print(concatenacion2, type(concatenacion2))
```

**NOTA**: ¿Ves por qué los decimales en Python siempre van con puntos y no con comas? Con las colecciones el intérprete de Python se volvería loco.

Podemos ver tambien el tipo de la lista

```python
print(type(listas))
```

Calcular la longitud de la misma mediante el método *built-in* ya visto: `len()`

```python
len(listas)
```

Accedemos a los elemenos de la lista mediante corchetes `[]`

**Importante**. El primer elemento es el 0

```python
# También se puede saber la posición exacta de un elemento dentro de la lista
lista_ejemplo = ["Esto", 2, 3, "patata", "nuevo elemento"]
print(lista_ejemplo[0]) # 0 es el primer elemento en Python
print(lista_ejemplo[-1]) # -1 es el último elemento de la lista
print(lista_ejemplo[-2]) # si le vas restando núm, pasas al penúltimo, antepenúltimo, etc.
print(lista_ejemplo[2]) # En positivo también, solo recordar que "2" es para Python el "3"
print(lista_ejemplo[8]) # Si metemos un elemento fuera de rango sale el error correspondiente
```

### Metodos en Listas
Para el tipo de objeto lista, también hay una serie de métodos catacterísticos que nos permiten operar con ellas: añadir valores, quitarlos, indexado, filtrado, etc... En [este enlace](https://www.w3schools.com/python/python_ref_list.asp) puedes encontrar todos los métodos que podrás usar con listas.

```python
# a tener en cuenta que en un string los métodos no modificaban los objetos, pero en listas sí pueden hacerlo

asignaturas = ["Matemáticas", "Física", "Química", "Historia", "Lengua"]
print(asignaturas)

# para introducir un elemento al final de la lista
asignaturas.append("Filosofia")
print(asignaturas)

# para añadir un elemento en una posición determinada
asignaturas.insert(1, "Educacion Fisica")
print(asignaturas)

# para quitar elementos
# asignaturas.clear() , este elimina todos los elementos, deja la lista vacía

asignaturas.pop(1) # este elimina el elemento seleccionado en el índice(), que es la posición
print(asignaturas)

asignaturas.remove("Filosofia") # este elimina el elemento por su valor
print(asignaturas)

# Para ordenar la lista alfabéticamente:
asignaturas.sort()
print(asignaturas)

# Para ordenar la lista al revés:
asignaturas.reverse()
print(asignaturas)
```
