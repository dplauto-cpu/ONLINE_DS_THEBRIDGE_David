Para terminar con las diferentes posibilidades que ofrecen los argumentos de las funciones, vamos a tratar los argumentos clave-valor.  

Es decir, tenemos también la opción de introducir un diccionario como argumentos, de esta forma, aunque el numero de argumentos sea variable, tendremos un inidicador, la clave, y el valor de cada clave.

Se implementa con `**`

```python
#Se trata de introducir un diccionario como indicador, aunque se trate de argumentos tendremos una manera de\
#recuperarlos por clave y por valor, se implementa con doble **

def insertar_movil_bd(**clave_valor): #al haber metido aquí el doble ** transforma esto en diccionario
  print(clave_valor, type(clave_valor))

insertar_movil_bd(modelo = "S20", marca = "Samsung", camara = "48MPX")
```

```python
def insertar_movil_bd(**kwargs): #kwargs es la convención para llamar a los argumentos con **
  for clave,valor in kwargs.items():
    # insertar(clave,valor)
    print(clave, ":", valor)

insertar_movil_bd(modelo = "S20", marca = "Samsung", camara = "48MPX", flip = True) #además admite añadir nuevas etiquetas
```

### Combinar `*args` con `**kwargs`
[al indice](#Contenidos)  
No hay ningun problema en tener un numero variable de argumentos y también argumentos clave-valor, todo ello en la misma función.

```python
def insertar_movil_bd(*args, **kwargs):
  print(type(args))
  for i in args:
    print(i)



insertar_movil_bd("es muy bueno", "no dura lo suficiente la bateria")
```

```python
def insertar_movil_bd(*args, **kwargs): # **kwargs no está en la función pero no da error, porque puede aparecer o no
  print(type(args))
  for i in args:
    print(i)

  print(type(kwargs)) # como antes, ejecuta la función y se pueden añadir elementos
  for clave,valor in kwargs.items():
    print(clave,valor)

insertar_movil_bd("es muy bueno", "no dura lo suficiente la bateria", marca = "Samsung", modelo = "S20", precio =1100)
```

### Argumentos keyword
[al indice](#Contenidos)  
Existe otro tipo de argumentos que son los *keyword*. Se caracterizan porque llevan un valor por defecto, y por tanto, si no usamos dicho argumento en la llamada, dentro de la función tomará el valor que hayamos dejado por defecto.

Ten en cuenta que estos argumentos se colocan **al final**

```python
round(12.3564, ndigits = 2) #nigit lo ha sacado de los shift+tab, ndigit es un argumento keyword
```

```python
sorted([12,3,24,54,12], reverse = True) #sorted y reverse son argumentos keyword
```

```python
def pedido_online(pedido, fecha, incidencia = False): #lo más importante de los keyword es colocarlos al final del código
  if incidencia == True:
    print("pongase en contacto atención al cliente")
  print("Su pedido se entregará en fecha", fecha)

pedido_online("tv LG", "12/12/2023")
pedido_online("Imac 2327 pulgadas ''", "10/11/23", incidencia = True)
```
