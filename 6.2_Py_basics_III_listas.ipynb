### Introducción
[al indice](#Contenidos)  

Ya conocemos bastante las listas. Veamos un repaso de lo que podemos hacer con ellas, así como algunas funcionalidades nuevas

```python
# Listas de números, strings, booleanos, con elementos repetidos, listas anidadas...
nums = [6, 2, 8, 3, 4, 5, 5]
months = ["Enero", "Febrero", "Marzo"]
mix = [5, 7, "Abril", True, None, ["Blanco", "Negro"]]
nums_int = [int(i) for i in nums]
calculo = max(nums_int) + min(nums_int) # este calculo es cosa mía, fuera de la píldora
print(calculo)
```

### Acceso
[al indice](#Contenidos)  

Entre corchetes introducimos el indice del elemento al que queremos acceder

```python
months = ["Enero", "Febrero", "Marzo"]
print(months[0])
print(months[-2])
```

Si planteamos el problema al revés. Tenemos los valores de la lista y lo que queremos es obtener el índice de esos valores dentro de la lista

```python
months = ["Enero", "Febrero", "Marzo", "Marzo"]
print(months.index("Febrero"))
print(months.index("Marzo")) # index devuelve el primer encontrado que coincida con lo buscado
```

**Slicing**: usamos slicing para acceder a varios elementos seguidos de la lista

```python
degrees = [22, 34, 15, 26, 18, 22]
sub_degrees = degrees[1:3] # muestra el elemento anterior al escrito, funciona como range
print(sub_degrees)
sub_degrees =degrees[1:6] # hay diversas maneras de mostrar todo el contenido, indicando inicio y fin
print(sub_degrees)
sub_degrees1 =degrees[1:len(degrees)] # mostrando toda la longitud de la lista
print(sub_degrees1)
sub_degrees2 = degrees[1:] # indicando solo el inicio y dejando un espacio en blanco
print(sub_degrees2)
```

### Modificar elementos
[al indice](#Contenidos)  

Las listas son mutables, por lo que podremos modificarlas

```python
# Accediendo mediante el indice
months = ["Enero", "Febrero", "Marzo"]
months[1] = "Abril" # el núm del index indica el elemento de las lista a sustituir
print(months)
```

### Añadir elementos
[al indice](#Contenidos)  

Se añade al final de la lista si usamos `append`, o si queremos en un lugar concreto, mediante `insert`

```python
# Con append lo añadimos al final de la lista
motorcycles = ['honda', 'yamaha', 'suzuki']
motorcycles.append("kawasaki") # append añade un elemento al final de la lista
print(motorcycles)
motorcycles.insert(3, "bultaco") # insert añade un elemento en el indice indicado
print(motorcycles)
```

```python
# Es muy comun crear una lista vacia, y a lo largo del programa, ir añadiendo elementos
lista_vacia = [] # y cuando haga falta se van introduciendo elementos
lista_vacia.append("primera marca")
print(lista_vacia)
# Un montón de líneas de código
lista_vacia.insert(0, "siguiente marca")
print(lista_vacia)
```

### Eliminar elementos
[al indice](#Contenidos)  

Para eliminar elementos se usar `remove`. Si no existe, da error, así que cuidado con esta sentencia.

En ocasiones resulta útil quedarnos con el elemento eliminado. Para ello usamos `pop()`, que elimina el elemento que le indiquemos, y además devuelve ese elemento por lo que podremos guardarlo en una variable para usarlo después.

```python
cars = ["VW", "Seat", "BMW", "VW"]
cars.remove("VW") # elimina el primer elemento encontrado
print(cars)
```

```python
# Eliminar elementos por indice, y guarda ese valor en una variable
cars = ["VW", "Seat", "BMW", "VW"]
eliminado = cars.pop(1) # pop permite eliminar el elemento que queramos, con la ventaja de que se puede recuperar
print(cars)
print(eliminado)
```

### Métodos: ordenar, tamaño, invertir orden, ocurrencia
[al indice](#Contenidos)  

En este apartado veremos los métodos más útiles, pero podrás consultar el resto en [este enlace](https://www.w3schools.com/python/python_ref_list.asp)

```python
degrees = [22, 34, 15, 26, 18, 22] # empezamos con método para ordenar la lista
degrees.sort()
print(degrees) #la ordena de menor a mayor, ascedente
degrees.sort(reverse = True)
print(degrees) #la ordena en descente, de mayor a menor
print(len(degrees)) #da la longitud de la lista
degrees.reverse() #invierte el orden de la lista, se diferencia del sort(reverse = True) en que no ordena sino que invierte, \
#pone el último el primero, el penúltimo al segundo, etc., si la lista estuviera desordenada contunuaría desordenada.
print(degrees)
#ocurrencia, saber el num de veces que aparece un elemento en la lista
print(degrees.count(22)) #este elemento no modifica la muestra, al contrario que los otros
```
