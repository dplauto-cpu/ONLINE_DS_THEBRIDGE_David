### Introducción
[al indice](#Contenidos)  

Muy similares a las listas. Si en Notebooks anteriores definíamos a las listas como ordenadas y mutables, las tuplas son ordenadas e inmutables.

* **Inmutables**: una vez creada la tupla con sus elementos iniciales, no se puede modificar
* **Ordenadas**: podemos acceder a sus elementos a través del indice y reordenar la tupla según queramos

Las tuplas se suelen usar para pequeñas colecciones de datos que no van a cambiar a lo largo del programa, como es el caso de las constantes.

Si las listas se construían mediante corchetes `[]`, las tuplas lo hacen con los paréntesis `()`

```python
tupla_ejemplo = (3, "texto")
print(tupla_ejemplo, type(tupla_ejemplo))
```

```python
print("el primer elemento es", (tupla_ejemplo[0])) #para buscar elementos se usa el index []
print("el segundo elemento es", (tupla_ejemplo[1]))
```

Los paréntesis ya se usan para reordenar operaciones `5 * (3 + 4)`, por lo que hay que añadir una coma cuando definamos una tupla con un único elemento, sino Python lo interpretará como un número.

```python
mala_tupla = (24)
print(type(mala_tupla)) #esto es lo que ocurre sin la coma, no lo identifica como tupla
buena_tupla = (24,)
print(type(buena_tupla)) #con la "," sí que lo identifica correctamente
```

```python
tupla_anid = (1,2, 3,("a","b","c")) #en las tuplas también se pueden anidar otras tuplas, \
#de hecho se pueden anidar listas y viceversa
print(tupla_anid)
print(len(tupla_anid)) # como en las listas, cuenta lo anidado como un solo elemento
```

Cuidado que las tuplas son inmutables, y una vez creadas no las podrás modificar después

```python
print(tupla_anid[2]) # index para ver un elemento en conreto
print(tupla_anid[3][1]) # para ver los elementos de la anexada se usa un doble index, \
#el 3 se refiere a la primera tupla (corresponde a la anidada) y el 1 al segundo elemento de la anidada
```

```python
# tuplas son iterables, como las listas, se pueden usar bucles
for i in tupla_anid:
    print(i)
```

```python
#también se puede hacer slicing
tupla_bis = tupla_anid[2:] #si no sse pone nada después de los ":" se indica que vaya hasta el final
print(tupla_bis)
```

### Listas y tuplas
[al indice](#Contenidos)  

Podemos combinar listas y tuplas que no tendremos ningún problema, siempre y cuando respetemos las propiedades de cada tipo de dato.

```python
tupla_error = (1, 2, 3, 4)
tupla_error[3] = 23 #el error más común en tuplas es tratar de modificarlas
```

```python
#Las tuplas sí permiten la concatenación
tupla_1 = (1,2,3)
tupla_2 = (4,5,6)
print(tupla_1 + tupla_2) #las imprime porque lo interpreta como una nueva tupla
tupla_3 = tupla_1 + tupla_2 #es igual que si hubiera creado una nueva variable

tupla_1 = tupla_1 + tupla_2 #Así, permite reasignación, estrictamente no he modificado tupla_1
print(tupla_1)
```

```python
lista_ejemplo = ["fresas", ("naranjas", "limones"), "kiwis"] #la lista contiene una tupla, se puede modificar\
#lo que no sea tupla, y en la tupla en teoría no se puede modificar nada
tupla_ejemplo = ("fresas", ["naranjas", "limones"], "kiwis") #la tupla contiene una lista
print(lista_ejemplo, type(lista_ejemplo))
print(tupla_ejemplo, type(tupla_ejemplo))

tupla_ejemplo.insert(1, "pera")
```
