### Introducción
[al indice](#Contenidos)  

En este tipo de colecciones no solo podemos agrupar datos, sino que tenemos la posibilidad de establecer una relación entre los mismos, con la forma **clave: valor**.

La sintaxis  es mediante llaves `{}`

> nombre_diccionario = {clave1: valor1, clave2: valor2, clave3: valor3}

Si veíamos que las listas son elementos mutables y y ordenadas, y las tuplas inmutables y ordenadas, los diccionarios son:

* **Mutables**: podemos modificar sus elementos
* **Desordenado**: la potencia de acceso de los diccionarios es su formato clave-valor, no su orden. Aun así, el dicionario conserva el orden de los elemento según los vamos insertando.

Vamos a declarar un diccionario en el que definiremos elementos de un coche

```python
#la gran vpotenica del diccionario es que asigna una etiqueta o clave a cada valor, \
#se puede acceder ya no por indices sino por claves, formate{clave1:valor1, etc.}
```

Se puede almacenar todo tipo de datos: numero, cadena, booleano, listas, tuplas, diccionarios...

Cuando declaremos el diccionario, lo podemos hacer en una sola linea `coche = {"color"="Gris", "marca"...}`, aunque se recomienda por sencillez a la hora de leer el código, realizar esta acción en varias líneas. Si escribimos el primer elemento y presionamos enter, Jupyter introduce automáticamente la tabulación.

```python
coche_dict = { #el diccionario permite string, int, bool...
    "color":"rojo", #color, marca, modelo, peso y cambio son las claves, rojo, tesla [etc.] los valores
    "marca": "tesla",
    "modelo": "cybertruck",
    "peso" : 3500,
    "cambio automático" : True
}
coche_dict
```

### Acceso, modificación, añadir, eliminar
[al indice](#Contenidos)  



```python
# para acceder al diccionario
print(coche_dict["color"]) # escribir el nombre del diccionario y ["valor"] "" porque normalmente son strings
print(coche_dict["peso"])
# print(coche_dict["puertas"]) # error de buscar algo que no está en el diccionario

#para modificar valores
coche_dict["color"] = ["rojo", "gris"]
print(coche_dict)

#los diccionarios no tienen métodos como apppen o pop
#para AÑADIR una etiqueta
coche_dict["puertas"] = 6
print(coche_dict)

#ELIMINAR etiqueta
del coche_dict["puertas"] # poner algo entre corchetes o elimina todo el diccionario
print(coche_dict)

#Para CONOCER LAS CLAVES de un dicc.
print(coche_dict.keys())

#Método para conocer los valores
print(coche_dict.values())

#Método para conocer las claves y valores
print(coche_dict.items())
```

Para más detalle de los métodos de los diccionario puedes [consultar la documentación](https://www.w3schools.com/python/python_ref_dictionary.asp)

```python
# Podemos empezar tambien mediante un diccionario vacío
tienda = {} #también sirve tienda = Dict()

#y se van añadiendo elementos según hagan falta y avance la programación
tienda["direccion"] = "Montesa 12"
tienda["tamaño"] = 123
print(tienda)
```

Si hay una clave que no existe en el diccionario, saltará un error.

En este caso, el error es bastante intuitivo: `KeyError`, no encuentra la clave

```python
#el clasico es intentar acceder a un valor o clave que no existe
coche_dict["puertas"]
```

**¿Cómo solventar esto?** Sirmpre tienes la opción de usar el bloque `try/except`, pero en este caso, los diccionarios tienen una solución más elegante: `mi_diccionario.get(clave, valor si no existe)`

```python
#el método get es elegante porque si etiqueta no existe, crea una y le da un valor por defecto
numero_de_puertas = coche_dict.get("puertas", 4)
print(numero_de_puertas) #como no existe la clave, solo imprime el valor

#qué ocurre si la clave sí existe
marca = coche_dict.get("marca","desconocida")
print(marca) #sigue identificando tesla como valor, no lo sustituye por "desconocida"
```

### Iterar
[al indice](#Contenidos)  

Que el diccionario no esté ordenado, no quiere decir que no podamos iterar sus elementos, ya que también es un **interable**, como los son las listas o las tuplas.

```python
#las claves tb se ordenan cuando se crean
for i in coche_dict:
    print(i) #hemos iterado por claves
    print(coche_dict[i]) #aquí te dice la clave y el valor
```

```python
#las claves tb se ordenan cuando se crean
for i in coche_dict:
    print(i,":",coche_dict[i]) #el metodo anterior queda algo difícil de ver, este es mucho más ordenado
```

```python
coche_dict.items() #items te enseña el contenido, en realidad es una tupla de tuplas
```

```python
#el metodo más elegante para acceder al contenido de un dict es "items" + iterar en bucle
coche_dict.items() #y esto lo puedo iterar con un bucle
for clave, valor in coche_dict.items(): #donde clave, valor se puede buscar cualquier termino que sí exista,\
# items te devuelve el contenido completo
    print(clave,":",valor)

for color,rojo in coche_dict.items():
  print(color,":",rojo)
```
