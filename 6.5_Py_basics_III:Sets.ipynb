Otra colección *bulit-in* de Python, compuesta por un conjunto de *valores*. Muy parecidos a las listas. Tienen las siguientes características:
* **Mutables**: podemos modificarlos una vez se hayan creado
* **No ordenado**
* **Elementos unicos**: se compone de un conjunto de valores únicas

**¿Cuándo usar sets?** Cuando estemos buscando unicidad en nuestros datos y no nos importe el orden.

Te dejo por aquí [la documentación](https://docs.python.org/2/library/sets.html) para realizar consultas sobre los sets

Si sintaxis es:
>```Python
> mi_set = {elemento1, elemento2, elemento3}
>```

Tiene una sintaxis muy parecida a la de las listas, pero en este caso no se utilizan los `{}`.

````python
# un set es una colección no ordenada de elementos únicos, no repite elemento
colores = {"rojo", "verde", "azul", "naranja","azul"} #no incluye duplicados, cuando print los ignora
print(colores, type(colores))
print(len(colores)) #como no reconoce el sehundo azul dice 4
````

Los elementos del set son unicos, por lo que si en la declaración, o posteriormente añadiendo elementos hubiese algún duplicado, el set lo ignoraría. Es más, cuando aplicas el `len`, muestra la cantidad de valores únicos que tiene.

````python
#metodo para añadir valores
colores.add("magenta")
print(colores)

# si queremos añadir una lista, update
colores.update(["celeste", "burdeos", "naranja "])
print (colores)

#para quitar elementos #he silenciado este porque si no da error y no ejecuta el siguiente código
#colores.remove("naranja")
#print(colores)

#también son iterables
for color in colores:
    print(color)
````

Para comprobar si dos sets tienen los mismos elementos

````python
set1 = {1,2,4,4,2,1,2,1}
set2 = {2,1,4}

set1 == set2
````

````python
#Uno de los usos más utilizados de los sets es eliminar elementos duplicados de las listas
lista = [100,10,23,23,15,32,45,10,10,100, 23] #xa eliminar duplicados, se pasa de lista a set y de set a lista
print(lista)
print(set(lista)) #1 igual que cuando se cambia a int, float, str..., la palabra clave es set \
#este print ya no tiene los duplicados y tiene formato set
print(list(set(lista))) #aquí está la lista en formato lista sin duplicados
lista = list(set(lista)) #se hacen unas asignaciones para que la lista sea plenamente operativa
print(lista)
````
