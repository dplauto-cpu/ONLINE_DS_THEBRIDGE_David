Continuamos con las funciones definidas por el usuario. Recuerda que la sintaxis que vimos en el anterior notebook era:


> ```Python
> def nombre_funcion(entrada):
>    operaciones varias
>    return output
> ```

Hicimos un ejemplo de función que nos devolvía un resultado (output) que podíamos almacenar en una variable... pero las funciones **tampoco tienen por qué llevar un `return`**. No siempre es necesario un output. En tal caso, devuelve `None`

````python
from datetime import datetime

def que_hora_es():
    now = datetime.now().time()
    print(now) # no es necesario que lleve return, si usamos print obtenemos una función tb.
````

````python
hora = que_hora_es()
print(hora, type(hora)) #si lleva return, devuelve datetime, luego he cambiado el código de arriba por print(now)
````

````python
hora = que_hora_es()
print(hora, type(hora)) #devuelve None porque no lleva return, pero sí devuelve algo
````

````python
print(now) #now no existe más que en la función, fuera de ella da error
````

También **puedes poner varias salidas en el return**. En ese caso, si no se especifica nada más la salida de la función será de tipo "tupla" (tuple). Pero generalmente los agrupamos en una colección.

````python
def conversor_km_millas(distancia):
    millas = 0.62 * distancia
    metros = 1000 * distancia
    return [round(millas,1), millas, metros]
````

````python
conversor_km_millas(2221) #lo que ha salido es una tupla
````

````python
salida = conversor_km_millas(2221) #si le asignamos nombre sin el (2221) nos da class 'function'
print(salida, type(salida))
````

````python
salida = conversor_km_millas(2221) #y si modificamos la función para que el resultado sea una lista, pues sale class list
print(salida, type(salida))
````

### Tipos de datos de los argumentos
[al indice](#Contenidos)  

Lo que quieras: numeros, texto, listas, tuplas, diccionarios, objetos de clases que hayas definido...

````python
def recibe_mix(tupla, lista, diccionario):
  print("tupla_contiene")
  print(tupla)
  print("es de tipo", type(tupla))
  print("lista_contiene") #en lugar de copy/paste, es mejor un bucle, lo veremos en la sesión en vivo
  print(lista)
  print("es de tipo", type(lista))
  print("diccionario_contiene")
  print(diccionario)
  print("es de tipo", type(diccionario))
  return [type(tupla), (lista), (diccionario)]
````

````python
recibe_mix((12, 34, 23), ["esto","es","una","lista"], {"key1":"valor1"})
````

````python
recibe_mix(1,"hola",True) #esta celda demuestra que no se han definido los tipos, aunque la celda anterior fuera\
#correcta, no hemos forzado los tipos, solo son nombres.
````

````python
# Todo lo que declaremos dentro de la función se crea UNICAMENTE para la función
# Fuera de la misma, esas variables no existen
def km2millas(dist):
  millas = 0.62*dist
  return round(millas,2)

print(km2millas(200))
print(millas) #millas sigue sin existir fuera de la función
````

Se crea un namespace interno dentro de las funciones, es decir, que lo que declaremos dentro, se queda dentro. No lo podremos usar fuera. Además, ten en cuenta que todo lo que introduzcamos dentro de flujos de control (`if/else`, bucles...), nos vale para el resto de la función

````python
def numero_ifs(numero):
    if numero == 1:
        out = 1
    return out

numero_ifs(1)
````

````python
numero_ifs(2)
````

````python
# Si no introducimos argumentos en una función que SI tiene argumentos, salta un error de este estilo
millas = km2millas()

print(km2millas(200))
````
