Puedes implementar funciones con todos los argumentos que quieras. Ahora bien, ten en cuenta dos cosas:

1. **El orden** de los argumentos. Cuando llamemos a la función, tenemos que seguir el mismo orden de argumentos que en la declaración de la función.
2. **Son obligatorios**. Si los declaramos en la función, después al llamarla, tenemos que poner todos sus argumentos. Luego veremos que hay una manera de poner argumentos opcionales.

```python
def multiplica(x1,x2,x3,x4):
  resultado = (x1*x2*x3)/x4
  return resultado

print(multiplica(3,4,5,6)) #se llaman posicionales porque no le hemos dicho a qué corresponde cada valor, Py interpreta\
#que el primer valor se corresponde con la primera variable, si faltara alguno (x1,x2,x3) daría error.
print(multiplica(3,4,6,5))#y tab inluye el orden, por eso se llaman , el print se lo añadido para ejecute\
# ambas operaciones, si solo hay una no es necesario ponerlo. ver texto abajo
```

```python
multiplica(x1 =3, x4 =6, x3 =5, x2 = 4) #lo mejor xa no cometer errores es asignar un valor al nombre de las variables
```

Fijate que los argumentos siguen un determinado orden: x1, x2, x3, x4. Cuando llamamos a la función, introduciremos 4 argumentos y la función los recogerá en ese orden. Asignará 4 a x1, 6 a x2, etc. Podemos también especificar el nombre del argumento en la llamada, lo que nos permite tener mayor flexibilidad en el orden.

```python
multiplica (x4 = 0 , x1 = 3, x2 = 5, x3=12) #da un doble error, uno en la línea de ejecución, no se puede dividir por 0\
# pero el segundo hace referencia a la línea dos de la definición de la función
```

```python
#La forma de arreglarlo podría ser con try/except
try:
  multiplica(x4=0, x1=2, x3=6, x2=3)
except Exception as ex:
  print("error de la funcion")
  print(ex) #ex te indica el error
  '''
  Lo mejor para evitar estos errores es ir a la función y hacer una gestión\
  de excepciones o una comprobación de valores de variables
  '''
```

```python
def multiplica(x1,x2,x3,x4):
  if x4 == 0:
    print("No se puede dividir por 0, argumentox4 inválido")
  else:
    resultado = (x1*x2*x3)/x4
    return resultado
```

```python
multiplica(x4=0, x1=2, x3=6, x2=3)
```

Fijate que aparece toda la traza del error, tanto la línea donde llamas a la función, como el error dentro de la función. Podemos solventar el error, introduciendo un bloque `try/except`
