En los ejemplos de la sesión anterior teníamos que fijar un número concreto de argumentos para nuestras funciones, pero hay ocasiones que no tenemos seguro cuántos argumentos son. Por suerte, las funciones de Python nos aportan esa flexibilidad mediante `*`

Veamos cómo implementar una función multiplicadora con numero variable de argumentos

```python
def multiplica(*argumentos): #con el * delante lo interpreta como una tupla
  print(argumentos,type(argumentos))

multiplica(2,3,5)
multiplica(1,2,4,67,4)
```

```python
def multiplica(*args): #cuando queremos tener un núm variable de argumentos la convención en Py es "args"
  resultado = 1
  for elemento in args:
    resultado = resultado * 1
  print("El resultado es", resultado)
  return resultado

multiplica(2, 3, 5)
multiplica(23, 12)
multiplica()
```

```python
#La celda anterior da un error básico porque hay un 1 donde debería poner elemento. \
#Lo bueno es que incluso dejando () vacio
def multiplica(*args):
  resultado = 1
  for elemento in args:
        resultado = resultado * elemento
  print("el resultado es", resultado)
  return resultado

multiplica(2, 4, 5)
multiplica(23, 13)
multiplica() #incluso si está vacio recupera un resultado
```

```python
#y se puede jugar un poco más
def multiplica(*args):
  resultado = 1
  for indice, elemento in enumerate(args):
        print("el siguiente elemento es", args[indice])
        resultado = resultado * elemento
  print("el resultado es", resultado)
  return resultado

multiplica(2, 4, 5)
multiplica(23, 13)
multiplica() #incluso si está vacio recupera un resultado
```

Ten en cuenta que `*args` es algo variable con X elementos. Como no sabemos a priori cuantos son, tendremos que recorrerlos con un `for`, y para cada argumento, aplicarle una operación. Por tanto, `*args` es un iterable, en concreto una **tupla**. Lo que le está dando la funcionalidad de "argumentos variables" es `*`, no `args`. Igual que ponemos `*args`, podemos poner `*argumentos`.

Puedes combinar argumentos posicionales con los `*args`

```python
# En este ejemplo, uso el ultimo argumento para dividir todo lo que habiamos multiplicado por este argumento\
# el objeto es combinar argumentos
def multiplica_divide (*args, div):
  resultado = 1
  for i in args:
    resultado = resultado * i
  resultado = resultado / div
  return resultado

multiplica_divide(10,31,24,5) #sale error porque Py no puede entender el valor de div, ya que no se ha especificado
```

```python
def multiplica_divide (*args, div):
  resultado = 1
  for i in args:
    resultado = resultado * i
  resultado = resultado / div
  return resultado

multiplica_divide(10,31,24,div = 5) #el indicamos el valor de div sí ejecuta y da el resultado
```

```python
def multiplica_divide (div, *args): #otra solución sencilla es un cambio en el orden, hace que se ejecute correctamente,\
# ya que interpreta que el primer valor es el que divide al resto de los argumentos
  resultado = 1
  for i in args:
    resultado = resultado * i
  resultado = resultado / div
  return resultado

multiplica_divide(5,10,31,24)
```

```python
def multiplica_divide (div, *args): #y otro error común es definir dos veces
  resultado = 1
  for i in args:
    resultado = resultado * i
  resultado = resultado / div
  return resultado

multiplica_divide(10,31,24,div = 5) #porque interpreta que hay varios valores div, no sabe cuando empieza args
```
