## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
grados = float(input("grados a radianes:"))
formula_radianes = grados * (3.1416 / 180)
print("resultado", formula_radianes)
print(f"{grados} grados son {formula_radianes} radianes")


```

## Ejercicio 5
