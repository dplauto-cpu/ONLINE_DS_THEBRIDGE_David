Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"

```python
class Tienda:
    #Atributos comunes 
    tipo = "Electrodomésticos"
    abierta = True

    def __init__(self, nombre, direccion, num_empleados, ventas_3meses): 
        #Atributos únicos tienda
        self.nombre = nombre
        self.direccion = direccion
        self.num_empleados = num_empleados
        # ventas_3meses es [mes1, mes2, mes3]
        self.ventas_3meses = ventas_3meses

    def calcular_total_ventas(self):
        #Suma ventas últimos 3 meses.
        return sum(self.ventas_3meses)

    def media_ventas_por_empleado(self):
        #Media ventas por empleado.
        if self.num_empleados == 0:
            return 0
        total = self.calcular_total_ventas()
        return total / self.num_empleados

    def obtener_info(self):
        #Nombre y la dirección de la tienda.
        return f"Tienda: {self.nombre}, Dirección: {self.direccion}"

    def ventas_ultimo_mes(self):
        #"""ventas último mes (último índice de la lista).
        return self.ventas_3meses[-1]

    def proyeccion_marketing(self, inversion):
        """
        Estimación ventas
        X < 1000: aumenta 20% (x1.2)
        X >= 1000: aumenta 50% (x1.5)
        """
        if inversion < 1000:
            factor = 1.2
        else:
            factor = 1.5
            
        #Nuevo atributo de ventas
        self.ventas_3meses = [v * factor for v in self.ventas_3meses]
        
        return self.calcular_total_ventas()

# --- Ejemplo de uso ---
if __name__ == "__main__":
    # Crear una tienda
    Teletin = Tienda("Teletin", "Calle La Puebla", 5, [1000, 1500, 1200])

    print(Teletin.obtener_info())
    print(f"Total ventas €: {Teletin.calcular_total_ventas()}")
    print(f"Media por empleado €: {Teletin.media_ventas_por_empleado()}")
    print(f"Ventas último mes €: {Teletin.ventas_ultimo_mes()}")

    # Aplicar proyección de marketing (inversión >= 1000)
    nuevo_total = Teletin.proyeccion_marketing(1500)
    print(f"Ventas tras marketing: {nuevo_total}")
    print(f"Ventas actualizadas (3 meses): {Teletin.ventas_3meses}")

```
